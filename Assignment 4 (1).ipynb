```python
def bcci_match(url):
    
    #getting the url opened
    driver.get(url)
    
    #creating empty lists to store required data
    dates = []
    timing = []
    match = []
    place = []
    teams = []
    
    #getting the webpage opened
    but = driver.find_element_by_xpath("//a[@class='navigation__link navigation__link--in-drop-down']")
    driver.get(but.get_attribute('href'))
    time.sleep(5)
    #for dates
    try:
        mon = driver.find_elements_by_xpath("//span[@class='fixture__month']")
        dat = driver.find_elements_by_xpath("//span[@class='fixture__date']")
        for i in range(len(mon)):
            dates.append(mon[i].text + ' ' + dat[i].text)
            
    except StaleElementReferenceException:
        WebDriverWait(driver,30).until(EC.presence_of_element_located((By.CLASS_NAME,'fixture__month')))
        driver.refresh()
        mon = driver.find_elements_by_xpath("//span[@class='fixture__month']")
        dat = driver.find_elements_by_xpath("//span[@class='fixture__date']")
        for i in range(len(mon)):
            dates.append(mon[i].text + ' ' + dat[i].text)
            
    #for time
    try:
        tim = driver.find_elements_by_xpath("//span[@class='fixture__time']")
        for i in tim:
            timing.append(i.text)
    except StaleElementReferenceException:
        WebDriverWait(driver,30).until(EC.presence_of_element_located((By.CLASS_NAME,'fixture__month')))
        driver.refresh()
        time.sleep(5)
        tim = driver.find_elements_by_xpath("//span[@class='fixture__time']")
        for i in tim:
            timing.append(i.text)
            
    #for teams 
    team_list = driver.find_elements_by_xpath("//p[@class='fixture__team-name']")
    t1 = team_list[::2]
    t2 = team_list[1::2]
    for i in range(len(t1)):
        teams.append(t1[i].text +' v '+ t2[i].text)
        
    #for place
    std = driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span[1]")
    for i in std:
        place.append(i.text)
        
    #for match
    mat = driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']")
    for i in mat:
        match.append(i.text)
        
    #creating the dataframe
    df_cricket = pd.DataFrame({})
    df_cricket['Dates'] = dates
    df_cricket['Teams'] = teams
    df_cricket['Place'] = place
    df_cricket['Time'] = timing
    df_cricket['Matches'] = match
    
    return df_cricket
```

```python
bcci_match(url)
```

```python
def sel_guru(url):
    
    #getting the website
    driver.get(url)
    
    #creating the lists
    nam_list = []
    desc_list = []
    
    #for getting the required webpage
    web1 = driver.find_element_by_xpath("//a[@title='Selenium']")
    driver.get(web1.get_attribute('href'))
    
    #for getting the second webpage
    web2 = driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']")
    driver.get(web2.get_attribute('href'))
    
    #for name
    nam = driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr//td[1]")
    for i in nam:
        nam_list.append(i.text)
        
    #for description
    desc = driver.find_elements_by_xpath("//table[@class='table table-striped']/tbody/tr//td[2]")
    for i in desc:
        desc_list.append(i.text)
        
    #creating the dataframe
    df_exp_hand = pd.DataFrame({})
    df_exp_hand['Exception Name'] = nam_list[1:]
    df_exp_hand['Description'] = desc_list[1:]
    
    return df_exp_hand
```

```python
def stat_gdp(url):
    
    #getting the website started
    driver.get(url)
    
    #opening the required webpage
    clic = driver.find_element_by_xpath("//div[@class='dropdown-content']//a[3]")
    driver.get(clic.get_attribute('href'))
    
    #opening the second required webpage
    clic1 = driver.find_element_by_xpath("//ul[@style='list-style-type:none;margin-left:20px;']/li/a")
    driver.get(clic1.get_attribute('href'))
    
    #creating the empty lists
    rank1 = []
    rank2 = []
    name1 = []
    name2 = []
    gsdp20_1 = []
    gsdp19_1 = []
    gsdp20_2 = []
    gsdp19_2 = []
    share1 = []
    share2 = []
    gdp1 = []
    gdp2 = []
    
    #rank
    ran1 = driver.find_elements_by_xpath("//tr[@class='odd']//td[1]")
    ran2 = driver.find_elements_by_xpath("//tr[@class='even']//td[1]")
    for i in ran1:
        rank1.append(i.text)
    for i in ran2:
        rank2.append(i.text)
        
    #name
    nam1 = driver.find_elements_by_xpath("//tr[@class='odd']//td[2]")
    nam2 = driver.find_elements_by_xpath("//tr[@class='even']//td[2]")
    for i in nam1:
        name1.append(i.text)
    for i in nam2:
        name2.append(i.text)
        
    #GSDP 19-20
    gsdp1 = driver.find_elements_by_xpath("//tr[@class='odd']//td[3]")
    gsdp2 = driver.find_elements_by_xpath("//tr[@class='even']//td[3]")
    for i in gsdp1:
        gsdp20_1.append(i.text)
    for i in gsdp2:
        gsdp20_2.append(i.text)
        
    #GSDP 18-19
    gsdp3 = driver.find_elements_by_xpath("//tr[@class='odd']//td[4]")
    gsdp4 = driver.find_elements_by_xpath("//tr[@class='even']//td[4]")
    for i in gsdp3:
        gsdp19_1.append(i.text)
    for i in gsdp4:
        gsdp19_2.append(i.text)
        
    #share
    shar1 = driver.find_elements_by_xpath("//tr[@class='odd']//td[5]")
    shar2 = driver.find_elements_by_xpath("//tr[@class='even']//td[5]")
    for i in shar1:
        share1.append(i.text)
    for i in shar2:
        share2.append(i.text)
        
    #GDP
    gd1 = driver.find_elements_by_xpath("//tr[@class='odd']//td[6]")
    gd2 = driver.find_elements_by_xpath("//tr[@class='even']//td[6]")
    for i in gd1:
        gdp1.append(i.text)
    for i in gd2:
        gdp2.append(i.text)
        
    #creating the dataframe
    df_gdp1 = pd.DataFrame({})
    df_gdp2 = pd.DataFrame({})

    #For odd 
    df_gdp1['Rank'] = rank1
    df_gdp1['State'] = name1
    df_gdp1['GSDP 19-20'] = gsdp20_1
    df_gdp1['GSDP 18-19'] = gsdp19_1
    df_gdp1['Share'] = share1
    df_gdp1['GDP'] = gdp1

    #For even
    df_gdp2['Rank'] = rank2
    df_gdp2['State'] = name2
    df_gdp2['GSDP 19-20'] = gsdp20_2
    df_gdp2['GSDP 18-19'] = gsdp19_2
    df_gdp2['Share'] = share2
    df_gdp2['GDP'] = gdp2

    #For creating a single dataframe
    df_gdp = pd.concat([df_gdp1,df_gdp2])
    df_gdp.set_index('Rank')
    df_gdp.sort_index(axis=0,inplace=True)
    
    return df_gdp
```

```python
url = 'http://statisticstimes.com/'
stat_gdp(url)
```

```python
def git_trend(url):
    
    #getting the website started
    driver.get(url)
    
    #using the search bar for further search
    search = driver.find_element_by_xpath("//label[@class='form-control input-sm header-search-wrapper p-0 js-chromeless-input-container header-search-wrapper-jump-to position-relative d-flex flex-justify-between flex-items-center']")
    search.send_keys('trending')
    search.send_keys(Keys.RETURN)
    time.sleep(5)
    
    #creating the empty lists
    name = []
    description = []
    contribute = []
    language = []
    
    #repository name
    nam = driver.find_elements_by_xpath("//div[@class='f4 text-normal']//a")
    for i in nam:
        name.append(i.text)
        
    #description
    desc = driver.find_elements_by_xpath("//p[@class='mb-1']")
    for i in desc:
        description.append(i.text)
        
    #contributors
    contri = driver.find_elements_by_xpath("//a[@class='muted-link']")
    for i in contri:
        contribute.append(i.text)
        
    #language used
    lang = driver.find_elements_by_xpath("//div[@class='d-flex flex-wrap text-small text-gray']//div[2]/span//span[2]")
    for i in lang:
        language.append(i.text)
        
    #creating the dataframe
    df_git = pd.DataFrame({})
    df_git['Repository Name'] = name
    df_git['Repository Description'] = description
    df_git['Contributors'] = contribute
    
    return df_git
```

```python
url = 'https://github.com/'
git_trend(url)
```

```python
def bill_list(url):
    
    #opening the website
    driver.get(url)

    #creating the empty lists
    song_title = []
    artist_name = []
    last_week = []
    peak = []
    weeks_on_board = []
    
    #getting to the required page
    nxt = driver.find_element_by_xpath("//a[@class='pill-button text--left shadow--soft']")
    nxt.click()

    #song title
    son = driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
    for i in son:
        song_title.append(i.text)
        
    #artist name
    art = driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
    for i in art:
        artist_name.append(i.text)
        
    #last week rank
    lw_rank = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
    for i in lw_rank:
        last_week.append(i.text)
        
    #Peak
    pea1 = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
    for i in pea1:
        peak.append(i.text)
        
    #weeks in charts
    wk_chart = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
    for i in wk_chart:
        weeks_on_board.append(i.text)
        
    #creating the dataframe
    df_bill = pd.DataFrame({})
    df_bill['Song Title'] = song_title
    df_bill['Artist Name'] = artist_name
    df_bill['Last Week Rank'] = last_week
    df_bill['Peak Rank'] = peak
    df_bill['No. of Weeks on Board'] = weeks_on_board

    return df_bill
```

```python
url = 'https://www.billboard.com/'
bill_list(url)
```

```python
def naukri_data(search_option):
    
    #getting the website
    driver.get('https://www.naukri.com/')
    
    #getting to the required webpage
    search = driver.find_element_by_xpath("//div[@class='inpWrap']/input")
    search.send_keys(search_option)
    search.send_keys(Keys.RETURN)
    time.sleep(10)
    
    #creating empty lists
    name = []
    company = []
    designation = []
    location = []
    skill = []
    
    #name
    nam = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
    for i in nam:
        name.append(i.text)
        
    #company
    comp = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
    for i in comp:
        company.append(i.text)
        
    #location
    loc = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span[1]")
    for i in loc:
        location.append(i.text)
        
    #designation
    desg = driver.find_elements_by_xpath("//div[@class='job-description fs12 grey-text']")
    for i in desg:
        designation.append(i.text)
        
    #skill
    skil = driver.find_elements_by_xpath("//ul[@class='tags has-description']")
    for i in skil:
        skill.append(list(i.text))
        
    #creating the dataframe
    df_naukri = pd.DataFrame({})
    df_naukri['Name'] = name
    df_naukri['Company'] = company
    df_naukri['Location'] = location
    df_naukri['Designation'] = designation
    df_naukri['Skill'] = skill
    
    return df_naukri
```
